```python
def find_headers(qmd_path = './generated_talks.qmd', talk_nums = []):
    # get all of the header lines
    with open(qmd_path, 'r') as f:
        dat = [e for e in f.readlines() if re.match(r'#+', e)]
    headers = [[e.strip() for e in dat if re.findall(f'{tnum}:', e)][0] for tnum in talk_nums]
    headers = [(
        h
        # disallowed characters
        .replace('#', '')
        .replace(':', '')
        .replace('(', '')
        .replace(')', '')
        .replace('.', '')
        .replace('?', '')
        # format as part of address
        .strip()
        .lower()
        .replace(' ', '-')
        ) for h in headers]

    return   [f"{qmd_path.replace('qmd', 'html')}#{h}" for h in headers]

schedule_links = pd.concat([
    pd.DataFrame({
        'Type': 'Talk', 
        'Name': Talks['POSTER'].tolist(),
        'Header': find_headers(qmd_path = './generated_talks.qmd', talk_nums = Talks['POSTER'].tolist())
        }),
        pd.DataFrame({
        'Type': 'Poster', 
        'Name': Posters['POSTER'].tolist(),
        'Header': find_headers(qmd_path = './generated_poster_list.qmd', talk_nums = Posters['POSTER'].tolist())
        }),
        pd.DataFrame({
        'Type': 'Keynote', 
        'Name': Keynotes['POSTER'].tolist(),
        'Header': find_headers(qmd_path = './generated_keynote_list.qmd', talk_nums = Keynotes['POSTER'].tolist())
        }),
        pd.DataFrame({
        'Type': 'Lightning', 
        'Name': Lightning['POSTER'].tolist(),
        'Header': find_headers(qmd_path = './generated_lightning_list.qmd', talk_nums = Lightning['POSTER'].tolist())
        })
        ]).reset_index(drop = True)


schedule_links.to_csv('./generated/schedule_links.csv', index=None)
```
